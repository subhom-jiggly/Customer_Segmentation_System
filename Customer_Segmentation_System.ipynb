```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
data = pd.read_csv('Mall_Customers.csv')

# Select the features for clustering (Age, Annual Income, Spending Score)
X = data[['Age', 'Annual Income', 'Spending Score']]

# Perform K-Means clustering
kmeans = KMeans(n_clusters=6)  # You can change the number of clusters as needed
kmeans.fit(X)

# Add the cluster labels to the DataFrame
data['Cluster'] = kmeans.labels_

# Visualize the clusters (2D plot for Annual Income vs. Spending Score)
plt.scatter(data['Annual Income'], data['Spending Score'], c=data['Cluster'], cmap='rainbow')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Customer Segmentation')
plt.show()

# Print the first few rows of the DataFrame with cluster labels
print(data.head())
```
